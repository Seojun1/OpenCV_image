# 7. 이미지 자르기

영역을 잘라서 새로운 윈도우(창)에 표시

```python
import cv2
img = cv2.imread('img.jpg')
# img.shape (얼만큼 짜를지 정의해야되기 때문에 이미지의 기본 사이즈 측정) [390, 640, 3]

crop = img[100:200, 200:400] # 세로 기준 100~200까지, 가로 기준 200~400까지 자름

cv2.imshow('img', img)
cv2.imshow('crop', crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

영역을 잘라서 기존 윈도우에 표시

```python
import cv2
img = cv2.imread('img.jpg')

crop = img[100:200, 200:400] # 세로 기준 100~200까지, 가로 기준 200~400까지 자름
img[100:200, 400:600] = crop # 자른 이미지를 어디에 위치할지 지정해줌

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
